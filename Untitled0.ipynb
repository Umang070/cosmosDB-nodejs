```python
import pandas as pd
```

```python
profiles = pd.read_csv("/content/insta_profiles.csv")
profiles.head()
```

```python
locations = pd.read_csv("/content/insta_locations.csv")
locations.rename({'id': 'location_id'}, axis=1, inplace=True)
locations.head()
```

```python
posts = pd.read_csv("/content/drive/MyDrive/instagram_posts.csv",sep="\t",nrows=5000000,skiprows=5000000,header=None)
posts.head()
```

```python
posts.dropna(subset=[posts.columns[3],posts.columns[4]], how='any', inplace=True)
```

```python
posts.drop([posts.columns[0],posts.columns[1],posts.columns[7]], axis=1, inplace=True)
```

```python
posts = posts.astype({posts.columns[1]:'int64',posts.columns[2]:'int64'})
```

```python
posts.head()
```

```python
# insta_posts =  pd.merge(pd.merge(posts,profiles,on='profile_id'),locations,on='location_id')
insta_posts = pd.merge(posts,profiles,left_on=posts.columns[1],right_on='profile_id').merge(locations,left_on=posts.columns[2], right_on='location_id')
insta_posts.head()
```

```python
insta_posts.drop([insta_posts.columns[1],insta_posts.columns[2],insta_posts.columns[7],insta_posts.columns[14],insta_posts.columns[12]], axis=1, inplace=True)
```

```python
insta_posts.head()
```

```python
insta_posts.rename({2: 'post_id', 5: 'cts_x', 6: 'post_type', 8: 'number_likes', 9: 'number_comments'}, axis=1, inplace=True)
```

```python
insta_posts.to_json("modifiedSecond_chunk.json",orient='records')
```
